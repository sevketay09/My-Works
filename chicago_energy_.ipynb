```
X=ndata.iloc[:, [0,14,15,30,32,33,35,36,37,38,39,40,41,42,43]].values
y=ndata.iloc[:, 2].values
y
```

```
from sklearn.model_selection import train_test_split
#split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

# **MinMax Scaler**

```
#Her özelliği belirli bir aralığa ölçeklendirerek özellikleri dönüştürür.
#Bu tahmin edici, her bir özelliği bireysel olarak, eğitim setinde verilen aralıkta olacak şekilde ölçeklendirir ve çevirir; sıfır ile bir arasında.

from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test) 
```

# **Linear Regression**

```
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(X_train, y_train)

#X_test'ime göre y_pred yani x_testime göre tahminleme yapıyorum.
y_pred = mlr.predict(X_test)
from sklearn.metrics import mean_squared_error
from math import sqrt

rmse = sqrt(mean_squared_error(y_test, y_pred))

from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)

adj_r2 = 1 - float(len(y)-1)/(len(y)-len(mlr.coef_)-1)*(1 - r2)

rmse, r2, adj_r2
```

# **KNN**

Bizim bir tane değerlendirme metriğimiz var, yani modelimizi değerlendireceğimiz sayısal bir sonuç;
Buna R^2 denir.Ve bu R'nin karesini almak demek değildir. r_square ' dir adı. 
R^2 = 1- (SSR/SST)
Eğer R^2 değeri bire ne kadar yakın ise bu o kadar iyi demektir!!

```
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=10, weights='distance')
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

from sklearn.metrics import mean_squared_error
from math import sqrt
sqrt(mean_squared_error(y_test, y_pred))
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)

adj_r2 = 1 - float(len(y)-1)/(len(y)-len(mlr.coef_)-1)*(1 - r2)

r2, adj_r2
```
